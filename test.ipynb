```python
import numpy as np
import matplotlib.pyplot as plt
from GaussianProcess import GP
from material_model import MaterialSimulator as sim
from sklearn.preprocessing import StandardScaler

def _Standardize(X):
    scaler = StandardScaler()
    return scaler.fit_transform(X)

X_train = np.random.rand(100, 10)
X_train = _Standardize(X_train)
sim_model = sim(noise_std=0.02, random_state=42)
y_train = sim_model(X_train)
print("Training data shape:", X_train.shape, y_train.shape)

X_test = np.random.rand(100, 10)
X_test = _Standardize(X_test)
weights = [0.5, 0.3, 0.2]  # 聚合权重示例
gp = GP(length_scale=1.0, sigma_f=1.0, sigma_n=1e-8, agg_weights=weights)
gp.fit(X_train, y_train)
results = gp.predict(X_test)

query_idx = np.argsort(results['std_agg'])[-5:]
print("Top 5 query indices:", query_idx)
```
